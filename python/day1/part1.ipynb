```python
lines = []
with open('input.txt') as f:
    lines = f.readlines()
    print(lines)
```

```python
values = []
for line in lines:
    val = ""
    for chr in line:
        if chr.isdigit():
            val += chr
    values.append(val)
print(values)
```

```python
finalists = []
for val in values:
    if len(val) == 2:
        finalists.append(int(val))
    elif len(val) == 1:
        finalists.append(int(val + val))
    else:
        finalists.append(int(val[0] + val[-1]))
print(finalists)
```
