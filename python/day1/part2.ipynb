```python
lines = []
with open('input.txt') as f:
    lines = f.readlines()
    print(lines)
```

```python
digit_words = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
overlaps = {
    "oneight": "18",
    "twone": "21",
    "threeight": "38",
    "fiveight": "58",
    "sevenine": "79",
    "eightwo": "82",
    "eighthree": "83",
    "nineight": "98"
}

corrected_lines = []
for line in lines:
    for word in digit_words:
        for ovr in overlaps.keys():
            line = line.replace(ovr, overlaps[ovr])
        
        if word in line:
            line = line.replace(word, str(digit_words.index(word) + 1))
    corrected_lines.append(line)
print(corrected_lines)
```

```python
values = []
for line in corrected_lines:
    val = ""
    for chr in line:
        if chr.isdigit():
            val += chr
    values.append(val)
print(values)
```

```python
finalists = []
for val in values:
    if len(val) == 2:
        finalists.append(int(val))
    elif len(val) == 1:
        finalists.append(int(val + val))
    else:
        finalists.append(int(val[0] + val[-1]))
print(finalists)
```
