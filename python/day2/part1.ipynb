```python
lines = []
with open('input.txt', mode="r") as f:
    lines = f.readlines()
    print(lines)
```

```python
winning_id = set()
color_limit = {
    "red": 12,
    "green": 13,
    "blue": 14
}

color_score = {
    "red": 0,
    "green": 0,
    "blue": 0
}
for line in lines:
    fail = False
    game_id, _, combinations = line.partition(":") 
    game_id = game_id.split(" ")[1]
    rounds = combinations.split(";")
    
    # print(game_id, combinations)
    for x, round in enumerate(rounds):
        revealed_colors = [val.strip() for val in round.split(", ")]
        print(round, revealed_colors)
        for color_combination in revealed_colors:
            score, color = color_combination.split(" ")
            # print(score, color)
            # print(color_score[color])
            color_score[color] = int(score)
            print(f"color score is {int(score)} for color {color} on round {x + 1}")
        for color_metric in color_score.keys():
            if (color_score[color_metric] > color_limit[color_metric]):
                fail = True
    if not fail:
        winning_id.add(int(game_id))
    color_score = {
        "red": 0,
        "green": 0,
        "blue": 0
    }
print(color_score)
print(winning_id)
print(sum(winning_id))
```
