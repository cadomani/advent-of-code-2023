```python
lines = []
with open('input.txt', mode="r") as f:
    lines = f.readlines()
    print(lines)
```

```python
multiplied = []
color_highest = {
    "red": 0,
    "green": 0,
    "blue": 0
}
color_score = {
    "red": 0,
    "green": 0,
    "blue": 0
}
min_values = []
for line in lines:
    fail = False
    game_id, _, combinations = line.partition(":") 
    game_id = game_id.split(" ")[1]
    rounds = combinations.split(";")
    
    print(f"game_id is {game_id}")
    for x, round in enumerate(rounds):
        revealed_colors = [val.strip() for val in round.split(", ")]
        for color_combination in revealed_colors:
            score, color = color_combination.split(" ")
            color_score[color] = int(score)
        for color_metric in color_highest.keys():
            if (color_score[color_metric] > color_highest[color_metric]):
                color_highest[color_metric] = color_score[color_metric]

    min_values.append(color_highest)
    mult = color_highest["red"] * color_highest["blue"] * color_highest["green"]
    multiplied.append(mult)
    
    color_highest = {
    "red": 0,
    "green": 0,
    "blue": 0
    }
    color_score = {
        "red": 0,
        "green": 0,
        "blue": 0
    }
    
    print(color_highest)
    print(f"{mult=}")
    

print(sum(multiplied))
```
